```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from transformers import pipeline

# Charger les données
df = pd.read_csv('../data/Emotion_final.csv')


# Créer une correspondance entre les labels de votre jeu de données et ceux du modèle
label_mapping = {
    'happy': 'joy',
    'sadness': 'sadness',
    'anger': 'anger',
    'fear': 'fear',
    'surprise': 'surprise'
}

# Remplacer les labels dans votre jeu de données
df['Emotion'] = df['Emotion'].map(label_mapping)

# Vérifier les valeurs uniques pour s'assurer que le remplacement a fonctionné
print(df['Emotion'].unique())

# Séparer les données en ensembles d'entraînement et de test
train_texts, test_texts, train_labels, test_labels = train_test_split(df['Text'], df['Emotion'], test_size=0.2)

# Créer une pipeline de classification
classifier = pipeline("sentiment-analysis", model="michellejieli/emotion_text_classifier")

# Préparer les données de test
test_texts = test_texts.tolist()

# Utiliser la pipeline pour classer les textes de test
results = classifier(test_texts)

# Extraire les labels prédits
predicted_labels = [result['label'] for result in results]

# Créer un DataFrame avec les phrases, les émotions prédites et les émotions réelles
df_results = pd.DataFrame({
    'text': test_texts,
    'predicted_emotion': predicted_labels,
    'real_emotion': test_labels.tolist()  # Convertir en liste
})

df_results
```

```python
# voir les emotions qui ont neutral
df_results[df_results['predicted_emotion'] == 'joy']
```

```python
# Calculer l'exactitude des prédictions
accuracy = accuracy_score(df_results['real_emotion'], df_results['predicted_emotion'])
print('Accuracy:', accuracy)
```
